##LEGUME CROYANCES ALLERGENES BOISSON REGISTRE

```python
def generate_legume_data(df_aliment, n=10):
    # Legume = (#nomAliment)
    available_aliments = df_aliment['nomAliment'].unique().tolist() # Les imports se font comme ceci : clé étrangère = df de la table source [ clé primaire ].unique.tolist()
    n_to_gen = min(n, len(available_aliments))
    chosen = random.sample(available_aliments, n_to_gen)
    return pd.DataFrame({'nomAliment': chosen})

def generate_croyances_data(df_aliment, n=5):
    # Croyances = (nomCroyance VARCHAR2(25), #(#nomAliment))
    croyances_types = ['Anti-ananas-sur-raclette','Anti-légume']
    available_aliments = df_aliment['nomAliment'].unique().tolist()
    data = []
    for _ in range(n):
        data.append({
            'nomCroyance': random.choice(croyances_types), # Selectionne une des deux croyances
            'nomAliment': random.choice(available_aliments) # puis le legume qui affecte la croyance
        })
    return pd.DataFrame(data)

def generate_allergenes_data(df_aliment, n=5):
    # Allergenes = (nomAllergie VARCHAR2(40), #(#nomAliment))
    allergies = ['Gluten', 'Lactose', 'Arachides', 'Fruits à coque', 'Soja']
    available_aliments = df_aliment['nomAliment'].unique().tolist()
    data = []
    for _ in range(n):
        data.append({
            'nomAllergie': random.choice(allergies),
            'nomAliment': random.choice(available_aliments)
        })
    return pd.DataFrame(data)

def generate_boisson_data():
    # Boisson = (nomBoisson VARCHAR2(50), type VARCHAR2(25))
    boissons = [
        ('Eau Cristaline', 'Eau'), ('Coca-Cola', 'Soda'), ('Fanta', 'Soda'),
        ('Vin de Savoie', 'Alcool'), ('Bière Artisanale', 'Alcool'),
        ('Jus d\'orange', 'Jus'), ('Thé Glacé', 'Thé')]
    # Ici on a essayé de prendre la liste entiere de boissons qu'on a créé
    df = pd.DataFrame(boissons, columns=['nomBoisson', 'type'])
    return df.reset_index(drop=True)

def generate_registre_data(n=10):
    # Registre = (idRegistre NUMBER(10))
    data = [{'idRegistre': fake.unique.random_number(digits=10, fix_len=True)} for _ in range(n)]
    return pd.DataFrame(data)

# Génération
df_legume = generate_legume_data(df_aliment, 10)
df_croyances = generate_croyances_data(df_aliment, 1000)
df_allergenes = generate_allergenes_data(df_aliment, 5)
df_boisson = generate_boisson_data()
df_registre = generate_registre_data(10)

print("Tables générées : Legume, Croyances, Allergenes, Boisson, Registre")
display(df_croyances.head())
```

##TENRAC

```python
def generate_tenrac_data(df_grade, df_rang, df_dignite, df_titre, n=2000):
    data = []

    # extrait des id des titre/rangs..
    grade_ids = sorted(df_grade['idGrade'].unique().tolist())
    rang_ids = df_rang['idRang'].unique().tolist()
    dignite_ids = df_dignite['idDignite'].unique().tolist()
    titre_ids = df_titre['idTitre'].unique().tolist()

    # pourcentage de personnes qui auront des grades élevés
    weights = []
    other_count = len(grade_ids) - 2
    remaining_prob = (1.0 - 0.6 - 0.01) / other_count if other_count > 0 else 0

    for gid in grade_ids:
        if gid == 1:
            weights.append(0.6)
        elif gid == 6:
            weights.append(0.01)
        else:
            weights.append(remaining_prob)

    weights = np.array(weights)
    weights /= weights.sum()

    for _ in range(n):
        nom = fake.last_name()[:25] # faux noms
        prenom = fake.first_name()[:25] # faux prenoms
        email = f"{nom.lower()}_{prenom.lower()}@{fake.free_email_domain()}"[:50] # email qui utilise nom_prenom

        # La presence d'un grade elevé augmente la presence d'un titre une dignité ou d'un rang
        selected_grade = np.random.choice(grade_ids, p=weights)
        presence_prob = 0.7 if selected_grade > 3 else 0.1

        id_rang = random.choice(rang_ids) if random.random() < presence_prob else None
        id_dignite = random.choice(dignite_ids) if random.random() < presence_prob else None
        id_titre = random.choice(titre_ids) if random.random() < presence_prob else None

        record = {
            'RFID': fake.unique.random_number(digits=10, fix_len=True),
            'nom': nom,
            'prenom': prenom,
            'courriel': email,
            'tel': ''.join([str(random.randint(0, 9)) for _ in range(10)]),
            'adresse_postale': fake.random_int(min=10000, max=99999),
            'dateDeNaissance': fake.date_of_birth(minimum_age=18, maximum_age=90),
            'idGrade': selected_grade,
            'idRang': id_rang,
            'idDignite': id_dignite,
            'idTitre': id_titre
        }
        data.append(record)

    return pd.DataFrame(data)

df_tenrac = generate_tenrac_data(df_grade, df_rang, df_dignite, df_titre, n=10000) # on en genere pleins !!
display(df_tenrac.head())
```

##REUNION


```python
from datetime import datetime, timedelta
import random

def generate_reunion_data(df_restaurant, df_repas, n_samples=150):
    # On definit quand commencent et finissent les reunions
    start_date = datetime(2020, 1, 1)
    end_date = datetime(2026, 3, 31)
    delta_days = (end_date - start_date).days
    all_dates = [start_date + timedelta(days=i) for i in range(delta_days + 1)]

    # Les probabilités doivent changer selon plusieurs paramètres :
    weights = []
    cutoff_date = datetime(2024, 3, 10) # La date de l'introduction aux légumes
    for d in all_dates:
        if d.month in [12, 1, 2]: # Les mois en hiver
            weight = 3.0 # proba haute
        elif d.month in [6, 7, 8]: # Les mois en été
            weight = 0.5 # probabilité basse
        else:
            weight = 1.0 # proba normale pour le reste
        if d > cutoff_date: # Après l'into aux legumes, probabilité extremement reduite
            weight *= 0.1
        weights.append(weight)
    weights = np.array(weights) / sum(weights)

    # Recupere la clé primaire de repas et vérifie que tout soit bien ajouté
    repas_ids = df_repas['idRepas'].unique().tolist()

    # recupere les adresses
    available_adresses = df_restaurant['adresse'].unique().tolist()
    reunion_records = []
    seen_pairs = set()

    attempts = 0
    while len(reunion_records) < n_samples and attempts < n_samples * 10:
        chosen_date = np.random.choice(all_dates, p=weights).date()
        chosen_addr = random.choice(available_adresses)

        if (chosen_date, chosen_addr) not in seen_pairs:
            seen_pairs.add((chosen_date, chosen_addr))
            reunion_records.append({
                'dateReunion': chosen_date,
                'adresse': chosen_addr,
                'idRepas': random.choice(repas_ids) # Clé étrangère vers Repas
            })
        attempts += 1

    return pd.DataFrame(reunion_records)

df_reunion = generate_reunion_data(df_restaurant, df_repas, n_samples=5000)
display(df_reunion.head())
```

```python
def generate_entretien_data(df_registre, df_machine, df_tenrac, n=15):
    data = []
    # clés etrangères
    registre_ids = df_registre['idRegistre'].tolist()
    machine_ids = df_machine['idMachine'].tolist()

    # triplette des clés primaires de tenrac
    df_tenrac_full = df_tenrac.merge(df_groupe, on='idGroupe')
    group_to_siret = {gid: random.choice(df_organisme['SIRET'].tolist()) for gid in df_groupe['idGroupe'].unique()}

    entretien_types = ['Préventif', 'Curatif', 'Installation', 'Mise à jour']

    for i in range(n):
        # on prends un tenrac
        tenrac_row = df_tenrac_full.sample(1).iloc[0]
        siret = group_to_siret[tenrac_row['idGroupe']]

        record = {
            'idEntretien': fake.unique.random_number(digits=10, fix_len=True), # NUMBER(10)
            'dateE': fake.date_between(start_date='-2y', end_date='now'),      # DATE
            'periodicite': random.choice([3, 6, 12, 24]),                    # NUMBER(2,0)
            'type': random.choice(entretien_types),                          # VARCHAR2(100)
            'idRegistre': random.choice(registre_ids),                       # FK Registre
            'idMachine': random.choice(machine_ids),                         # FK Machine
            'SIRET': siret,                                                   # FK Part (Organisme)
            'idGroupe': tenrac_row['idGroupe'],                               # FK Part (Groupe)
            'RFID': tenrac_row['RFID']                                        # FK Part (Tenrac)
        }
        data.append(record)
    return pd.DataFrame(data)

df_entretien = generate_entretien_data(df_registre, df_machine, df_tenrac, n=15)
print("Table Entretien générée avec succès.")
display(df_entretien.head())
```

```python
def generate_total_coverage_pairs(list_a, list_b, target_n):
    pairs = set()
    # TOUT LES ELEMENTS DE LA LISTE seront utilisés
    for a in list_a:
        pairs.add((a, random.choice(list_b)))

    # + des paires randoms
    all_possible = [(a, b) for a in list_a for b in list_b]
    random.shuffle(all_possible)

    for p in all_possible:
        if len(pairs) >= target_n:
            break
        pairs.add(p)

    return list(pairs)

# 1. df_participe:
participe_pairs_raw = generate_total_coverage_pairs(reunion_pairs, tenrac_triplets, 4000)
data_participe = [{'SIRET': s, 'idGroupe': g, 'RFID': r, 'dateReunion': dt, 'adresse': addr}
                  for (dt, addr), (s, g, r) in participe_pairs_raw]
df_participe = pd.DataFrame(data_participe)

# 2. df_utilise:
utilise_pairs = generate_total_coverage_pairs(repas_ids, machine_ids, 2000)
df_utilise = pd.DataFrame(utilise_pairs, columns=['idRepas', 'idMachine'])

# 3. df_crois:
crois_pairs_raw = generate_total_coverage_pairs(tenrac_triplets, croyance_names, 6000)
data_crois = [{'SIRET': s, 'idGroupe': g, 'RFID': r, 'nomCroyance': c} for (s, g, r), c in crois_pairs_raw]
df_crois = pd.DataFrame(data_crois)

# 4. df_sauce_constitue_de:
sauce_const_pairs = generate_total_coverage_pairs(sauce_names, aliment_names, 100)
df_sauce_constitue_de = pd.DataFrame(sauce_const_pairs, columns=['nomSauce', 'nomAliment'])

# 5. df_plat_constitue_de:
plat_const_pairs = generate_total_coverage_pairs(plat_ids, aliment_names, 1000)
df_plat_constitue_de = pd.DataFrame(plat_const_pairs, columns=['idPlat', 'nomAliment'])
```

```python
veggie_list = df_legume['nomAliment'].unique().tolist()
cutoff_date = pd.Timestamp('2024-03-10').date()

# 2. Identifier les menus 'pollues' et les menus 'sains'
df_all_menus = df_menu.merge(df_plat_constitue_de, on='idPlat')

# Menus avec légumes
bad_menu_ids = df_all_menus[df_all_menus['nomAliment'].isin(veggie_list)]['idMenu'].unique().tolist()

# les ids sont triés entre bons et mauvais menus ici
bad_repas_ids = df_repas[df_repas['idMenu'].isin(bad_menu_ids)]['idRepas'].unique().tolist()
good_repas_ids = df_repas[~df_repas['idRepas'].isin(bad_repas_ids)]['idRepas'].unique().tolist()

if not good_repas_ids:
    print("Warning: good_repas_ids is empty. This implies df_repas is empty or has issues.")
    # Fallback: if df_repas is truly empty, this will still lead to issues, but handles empty good_repas_ids otherwise.
    good_repas_ids = df_repas['idRepas'].unique().tolist() if not df_repas.empty else []

# probabilité de 40% de légumes après le cutoff
def random_veggie_mix(row):
    if row['dateReunion'] >= cutoff_date:
        if bad_repas_ids and random.random() < 0.4:
            return random.choice(bad_repas_ids)
        elif good_repas_ids:
            return random.choice(good_repas_ids)
        else:
            print("Error: No repas (good or bad) available for assignment. Returning original idRepas.")
            return row['idRepas']
    return row['idRepas']

df_reunion['idRepas'] = df_reunion.apply(random_veggie_mix, axis=1)
```

```python
veggie_set = set(df_legume['nomAliment'].unique())
cutoff_date = pd.Timestamp('2024-03-10').date()

df_full = df_reunion.merge(df_repas[['idRepas', 'idMenu']], on='idRepas')\
                    .merge(df_menu[['idMenu', 'idPlat']], on='idMenu')\
                    .merge(df_plat_constitue_de, on='idPlat')

df_full['is_bad_veggie'] = (df_full['nomAliment'].isin(veggie_set)) & (df_full['dateReunion'] >= cutoff_date)
veggie_reunions = df_full[df_full['is_bad_veggie']].groupby(['dateReunion', 'adresse'])['nomAliment'].first().to_dict()

avis_records = []
tenrac_rfids = df_tenrac['RFID'].tolist()

for _, reunion in df_reunion.iterrows():
    key = (reunion['dateReunion'], reunion['adresse'])
    has_veggies = key in veggie_reunions
    num = random.randint(2, 5)
    participants = random.sample(tenrac_rfids, num)

    for rfid in participants:
        if has_veggies:
            note, comm = random.randint(1, 2), f"Decevant, il y avait des {veggie_reunions[key]}..."
        else:
            note, comm = random.randint(4, 5), "Superbe raclette traditionnelle !"

        date_not = pd.to_datetime(reunion['dateReunion']) + timedelta(days=random.randint(0, 2))
        avis_records.append({
            'date': reunion['dateReunion'], 'adresse': reunion['adresse'], 'RFID': rfid,
            'note': note, 'commentaire': comm, 'dateNotation': date_not.date()
        })

df_avis = pd.DataFrame(avis_records)
display(df_avis)
```
